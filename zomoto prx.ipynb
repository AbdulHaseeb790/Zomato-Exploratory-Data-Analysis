```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"rating counts"})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="rating counts",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="rating counts",data=ratings,hue="Rating text",palette=['black','red','blue','green','purple','brown'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['black','red','blue','green','purple','brown'])
```

```python
sns.countplot(x="Country",data=final_df,hue="Aggregate rating")
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Arial'

# Seaborn style and context (set first)
sns.set_style("whitegrid")
sns.set_context("notebook")

# Create a new figure for each plot
plt.figure(figsize=(10,6))   # NEW figure

# Plot
sns.countplot(x="Currency", data=final_df, hue="Country")

# Title
plt.title("Currency Count by Country")

# Show plot and clear figure
plt.show()
```
